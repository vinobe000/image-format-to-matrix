```python
class lang_letter():
     lang_dict={}
     def __init__(self,lang_name,lang_location):
         import numpy as np
         from PIL import Image
         self.lang_name=lang_name
         self.lang_location=lang_location
         self.image=Image.open(self.lang_location)
         self.lang_letter_array=np.asarray(self.image)
         lang_letter.lang_dict.update({self.lang_name:self.lang_letter_array})
```

```python
A_lang_letters=lang_letter("Abaza","/content/drive/MyDrive/A Letter/Abaza.png")
```

```python
A_lang_letters=lang_letter("Abellen","/content/drive/MyDrive/A Letter/Abellen.gif")
```
